```r
#Tarih format değişikliği yapmak için
#R ın tarih formatı yıl ay gün şeklindedir.
trh = c("12/10/2001","25/01/2026")
#yıl ay gün olacak şekilde değiştirmek istiyoruz.
yeni_tarih = strptime(trh,format = "%d/%m/%Y")
yeni_tarih
```

**Veri Setini Alt Kümelere Ayırmak**

```r
isimler = c("mehmet","ahmet","özlem")
sinav_notu = c(75,85,96)
boy = c(175,185,168)
kg = c(75,85,56)
var_mi = c(T,F,F)

veri = as.data.frame(cbind(isimler,sinav_notu,boy,kg,var_mi))
veri
```

```r
#Genel tablodan kendi tablomuzu oluşturduk.
yeni_veri = veri[,c("isimler","kg")]
yeni_veri
```

```r
#satıra göre filtreleme yapar gibi de kendi tablomuzu oluşturabiliriz.
yeni_veri_2 = veri[veri$var_mi==F,]
yeni_veri_2
```

```r
#f olanların boylarını değer olarak öğrenebilirim.
yeni_veri_3 = veri[veri$var_mi==F,"boy"]
yeni_veri_3
class(yeni_veri_3)
```

```r
#veri setinden kendi filtremizi yaparak bir tablo daha oluşturduk subset() fonksiyonu ile
sorgu1 = subset(kadin, boy>125)
sorgu1
```

```r
#veri setinden çoklu filtreleme yaparak tablomuzu oluşturduk.
sorgu2 = subset(kadin, boy>125 & kilo<65)
sorgu2
```

```r
#boy u 125cm den büyük olan kadınların sadece kiloları gösteren kod
yeni_veri_4 = subset(kadin,boy>125, select = kilo)
yeni_veri_4
```
