# **Tarih ve Zaman**


```r
#iki tarih arasında ki fark.
#as.Date fonksiyonu ile string değeri tarihe dönüştürürüz.

ilk_tarih = as.Date("12/12/2003")
son_tarih = as.Date("15/02/2025")

fark = son_tarih-ilk_tarih
fark
```

```r
#sistem tarihini görmek için
#System.time()
```

```r
#iki tarih arasında ki fark
fark= difftime("12/10/2010","25/9/2025")
fark
class(fark)
```

```r
#Tarih format değişikliği yapmak için
#R ın tarih formatı yıl ay gün şeklindedir.
trh = c("12/10/2001","25/01/2026")
#yıl ay gün olacak şekilde değiştirmek istiyoruz.
yeni_tarih = strptime(trh,format = "%d/%m/%Y")
yeni_tarih
```

**Veri Setini Alt Kümelere Ayırmak**

```r
isimler = c("mehmet","ahmet","özlem")
sinav_notu = c(75,85,96)
boy = c(175,185,168)
kg = c(75,85,56)
var_mi = c(T,F,F)

veri = as.data.frame(cbind(isimler,sinav_notu,boy,kg,var_mi))
veri
```

```r
#Genel tablodan kendi tablomuzu oluşturduk.
yeni_veri = veri[,c("isimler","kg")]
yeni_veri
```
