# ** Listeler **

Listeler vektörlere benzer ancak vektörlerden farklı olarak herhangi bir veri türünde veri içerebilir. Esnek bir veri yapısıdır. Listelerde hem farklı türden, hem de farklı uzunlukta veriler yer alabilir. Listeler list() fonksiyonu ile oluşturulabilir.Vektöri matris ve diziyi bir arada tutabilir.

```r
# bir listeye koymak için üç ayrı vektör oluşturma
ad <- c("Ad1", "Ad2", "Ad3", "Ad4", "Ad5", "Ad6")
boy <- c(180, 181, 188, 170, 172, 175, 181)
kilo <- c(80, 85, 55)

# üç vektörden oluşan bir liste oluşturma
liste <- list(ad, boy, kilo)
liste
```

```r
#Listenin 1. elemanına ulaşmak için
liste[1]
```

```r
#Tek köşeli parantezde alt bileşenleri döndürülür iç içe köşeli pazantez ile ilgili alt bileşenin değerleri döndürülür.
liste[[1]]
```

```r
#Alt listenin istenilen elemanını görüntülemek için 2. bileşenin 3. elemanını verir 188 i verir

liste[[2]][3]
```

```r
bilgiler <- list(ad="Mehmet", yas= 40, aktif_ogrenci= TRUE)
bilgiler
```

```r
#Listenin istenilen elemanını çağırmak için
bilgiler[2]
```

```r
#Listenin istenilen elemanına değer atamak için
bilgiler[2]<-41
bilgiler
```

```r
#Listenin isimleri(KEY) yazdırmak istediğimde
names(bilgiler)
```

```r
#Bileşenleri değerleriyle birlikte yazdırmak için
unlist(bilgiler)
```

```r
# liste'nin yapısının incelenmesi
str(bilgiler)
```

```r
#Listeye yeni bir isim ve değer eklemek istediğimde
bilgiler$boy=178
str(bilgiler)
```

```r
#Listeden bir isim ve değer silinmek istendiğinde
bilgiler$boy=NULL
str(bilgiler)
```

```r
#Bir listenin seçilen isim değişkeninin değerini öğrenmek için
bilgiler$ad
```

```r
# karakterlerden ve sayılardan oluşan vektörlerin oluşturulması
personel <- c("Per01", "Per02", "Per03", "Per04", "Per05", "Per06", "Per07", "Per08",
              "Per09", "Per10")
motivasyon_puani <- c(90, 80, 80, 80, 70, 40, 70, 50, 90, 80)
maas <- c(40000, 40000, 30000, 35000, 20000, 12000, 18000, 25000, 50000, 30000)

# listenin list() fonksiyonu ile oluşturulması
liste1 <- list(personel, motivasyon_puani, maas)
liste1
```

```r
# liste2 nesnesinin oluşturulması için iki ayrı vektörün oluşturulması
sayilar <- 1:10
mantik <- c(T, F, F)
liste2 <- list(sayilar, mantik)

# liste3'e eklenecek olan harfler isimli vektörün oluşturulması
harfler <- letters[1:20]

# liste3'ün oluşturulması
liste3 <- list(liste1, liste2, harfler)
liste3
```
