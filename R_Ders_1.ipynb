# **ARİTMETİK OPERATÖRLER**

R'da temel aritmetik işlemleri yapmak için aşağıdaki operatörler kullanılır:

*   `+` : Toplama
*   `-` : Çıkarma
*   `*` : Çarpma
*   `/` : Bölme
*   `^` (veya `**`) : Üs alma
*   `%%` : Mod alma (kalanı bulma)
*   `%/%` : Tam sayı bölmesi (bölümün tam sayı kısmını bulma)

```r
sayi1 <-15
sayi2 <- 5
```

```r
#Toplama
sayi1+sayi2
```

```r
#Çıkarma
sayi1-sayi2
```

```r
#Bölme
sayi1/sayi2
```

```r
#Çarpma
sayi1*sayi2
```

```r
#Üs Alma
sayi1^sayi2
```

```r
#Mod Alma Kalanı Bulma
sayi1<-20
sayi2<-4

sayi1%%sayi2
```

```r
#Bölen Sayısını Bulma
sayi1%/%sayi2
```

```r
#Kare Alma
sayi2**2
```

```r
#Karekök Alma
sqrt(sayi2)
```

### **MANTIKSAL OPERATÖRLER**

R'da iki veya daha fazla koşulu birleştirmek veya tersine çevirmek için kullanılan mantıksal operatörler şunlardır:

*   `<` (BÜYÜKTÜR): koşul  doğruysa `TRUE` döndürür yanlışsa 'FALSE'.
*   `>` (KÜÇÜKTÜR): koşul  doğruysa `TRUE` döndürür.yanlışsa 'FALSE'.
*   `<=` (BÜYÜK EŞİT): koşul  doğruysa `TRUE` döndürür.yanlışsa 'FALSE'.
*   `>=` (KÜÇÜK EŞİT): koşul  doğruysa `TRUE` döndürür.yanlışsa 'FALSE'.
*   `&` (VE): Her iki koşul da doğruysa `TRUE` döndürür.yanlışsa 'FALSE'.
*   `|` (VEYA): Koşullardan en az biri doğruysa `TRUE` döndürür.
*   `!` (DEĞİL): Koşulun mantıksal tersini döndürür (doğruysa yanlış, yanlışsa doğru).
*   `&&` (Tek VE): İlk koşul yanlışsa ikinci koşulu değerlendirmez (tek elemanlı karşılaştırmalarda kullanılır).
*   `||` (Tek VEYA): İlk koşul doğruysa ikinci koşulu değerlendirmez (tek elemanlı karşılaştırmalarda kullanılır).
*   `==` (Eşit Mi): Eşit mi sorgusu yapılır. Eşitse TRUE döner.
*   `!=` (Eşit Değil Mi): Eşit değil  mi sorgusu yapılır. Eşit değilse TRUE döner.

```r
a <- 25
b <- 12
c <- -2
d <- 2.5
```

```r
a<b
```

a , b den küçük mü diye sorguladık. Yanıt bool bir değer olacak

```r
b <= a
```

```r
a<b | c<d
```

Yada ile  iki koşuldan bir tanesi TRUE ise TRUE döner.

```r
a==b
```

a , b ye eşit mi diye sorguladık.

```r
!a==b
```

! sonucu ters döndürür.
